Definizioni necessarie:

- Cosa si intende con arrivi (in ambito del turismo)
- Cosa si intende con presenze (in ambito del turismo)

1) **Arrivi**: numero di clienti, italiani e stranieri, ospitati negli esercizi ricettivi nel periodo considerato. <br>
2) **Presenze**: numero delle notti trascorse dai clienti, italiani e stranieri, negli esercizi ricettivi.

*Riferimenti*: Glossario Istat
https://www.istat.it/it/files/2011/12/glossario1.pdf

<br>

Arrivi: clienti o gruppi di clienti che arrivano negli esercizi <br>

Presenze: numero di notti trascorse dai clienti
(3 presenze -3 notti passate nell'esercizio- anche se viene considerato solo 1 arrivo)

<h1>Conta ed identificazione degli outliers</h1>

```python
# Funzione per identificare e contare gli outlier utilizzando l'IQR

# creo un nuovo dizionario che contiene solo le colonne che mi servono dei singoli df

data_categorie = {
    'Agriturismi Arrivi': arrivi_agriturismi,
    'Alberghi Arrivi': arrivi_alberghi,
    'Campeggi e Villaggi Arrivi': arrivi_campeggi,
    'Agriturismi Presenze': presenze_agriturismi,
    'Alberghi Presenze': presenze_alberghi,
    'Campeggi e Villaggi Presenze': presenze_campeggi,

}

# Funzione che conta quanti outliers ci sono
def count_outlier_iqr(data_categorie, titolo):
    data = data_categorie[titolo]
    # Calcola il 25° e il 75° percentile
    percentile_25 = np.percentile(data, 25)
    percentile_75 = np.percentile(data, 75)

    # Calcola l'IQR
    iqr = percentile_75 - percentile_25

    # limiti per gli outlier
    limite_inferiore = percentile_25 - 1.5 * iqr
    limite_superiore = percentile_75 + 1.5 * iqr

    # identifica gli outlier
    outlier_inferiori = data[data < limite_inferiore]
    outlier_superiori = data[data > limite_superiore]

    numero_outlier = len(outlier_inferiori) + len(outlier_superiori)
    return numero_outlier


# Calcola e conta gli outlier per ciascuna categoria di arrivi e presenze
for structure in data_categorie:
    numero_outlier = count_outlier_iqr(data_categorie, structure)
    print(f"Numero di outlier per {structure}: {numero_outlier} \n")

    # Stampa gli outlier inferiori e superiori
    data = data_categorie[structure]
    percentile_25 = np.percentile(data, 25)
    percentile_75 = np.percentile(data, 75)
    iqr = percentile_75 - percentile_25
    limite_inferiore = percentile_25 - 1.5 * iqr
    limite_superiore = percentile_75 + 1.5 * iqr

    outlier_inferiori = data[data < limite_inferiore]
    outlier_superiori = data[data > limite_superiore]

    print(f"Outlier inferiori per {structure}: {outlier_inferiori}")
    print(f"Outlier superiori per {structure}: {outlier_superiori}\n")
```

<h1>Agriturismi</h1>

```python
# Import delle librerie per il modello

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

# Definizione della variabile dipendente(y) e della variabile indipendente(x)
x = presenze_agriturismi.reshape(-1, 1)
y = arrivi_agriturismi.reshape(-1, 1)

# Split dei dati in set di addestramento e test
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)

# Creazione e addestramento del modello
model = LinearRegression()
model.fit(x_train, y_train)

# Predizione dei valori
y_pred = model.predict(x_test)

# Calcolo di R2 e MSE
r2 = r2_score(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)

print(f"R2 (Agriturismi): {r2:.4f} \n")

print(f"MSE (Agriturismi): {mse:.4f}")

# Visualizzazione dei valori predetti con valori reali

plt.figure(figsize=(12, 10))
plt.plot(np.arange(y_test.size), y_test, alpha=0.8, label="Valori Reali")
plt.plot(np.arange(y_pred.size), y_pred, alpha=0.8, label="Valori Predetti")
plt.title('Confronto tra Valori Reali e Predetti')
plt.xlabel('Campioni')
plt.ylabel('Variabile Dipendente (y)')
plt.legend()
plt.show()
```
